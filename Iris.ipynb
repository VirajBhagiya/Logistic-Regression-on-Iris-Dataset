Box Plot

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

plt1 = sns.boxplot(df['SepalLengthCm'], ax=axes[0,0])
plt2 = sns.boxplot(df['SepalWidthCm'], ax=axes[0,1])
plt1 = sns.boxplot(df['PetalLengthCm'], ax=axes[1,0])
plt2 = sns.boxplot(df['PetalWidthCm'], ax=axes[1,1])

plt.tight_layout()
```

```python
# Deal with outliers

Q1 = df.SepalWidthCm.quantile(0.25)
Q3 = df.SepalWidthCm.quantile(0.75)
IQR = Q3 - Q1
df = df[(df.SepalWidthCm >= Q1 - 1.5*IQR) & (df.SepalWidthCm <= Q3 + 1.5*IQR)]

plt.boxplot(df.SepalWidthCm)
```

```python
sns.boxplot(x = 'Species', y = 'SepalLengthCm', data = df)
```

```python
sns.displot(data=df, x="SepalLengthCm", hue="Species", kind="kde")
sns.displot(data=df, x="SepalWidthCm", hue="Species", kind="kde")
sns.displot(data=df, x="SepalLengthCm", hue="Species", kind="kde")
sns.displot(data=df, x="SepalWidthCm", hue="Species", kind="kde")
```

```python
sns.heatmap(df.corr(numeric_only=True), cmap='crest', annot=True)
```
